## theoretical

there are always multiple ways to do things in Python. Both numpy and scipy can generate random data according to different theoretical distributions

note, you will never really need these functions to do empirical work. We are going to simulate a bunch of fake data to get the feel for how these distributions work

### normal

normal has two parameters, the mean $\mu$ and the variance $\sigma^2$

not every distribution has those same two parameters, so numpy generalizes their names to *location* and *scale*

```python
rng = np.random.default_rng(seed=2026)
```

flip a coin 30 times, what's the probability you get 22 or more heads?

```python
bx = np.random.binomial(30, 0.5, 1000)
```

```python
bx
```

```python
sns.histplot(bx)
```

```python
bx[bx>=22]
```

```python
len(bx[bx>=22]) / 1000
```

Luka Doni shoots 78.5% from the free throw line. If he gets fouled 10 times in a game, what is the probability he makes 18 or more free throws (assuming all fouls are 2-point shots). One way to do this is with math. Another is by simulating the data

```python
# 20 shots, 78.5% make probability, simulated 1000 times
bx = np.random.binomial(20, 0.785, 1000)
```

```python
sns.histplot(bx, bins=50)
```

```python
len(bx[bx>=18]) / 1000
```

```python
scipy.stats.binomtest?
```

```python
scipy.stats.binomtest(18, 20, 0.785, alternative='greater') 
```
