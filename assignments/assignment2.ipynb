## Hypothesis Testing

you are a regional transportation planner studying traffic accidents. It seems there was a noticeable drop in accidents last year so you want to setup a formal test using data you've collected on accidents in 2024 and 2025. In a markdown cell below, write the null hypothesis and alternative hypothesis (in words, not math) you plan to test.



- **H0**: Traffic accidents in 2025 is greater than or equal to 2024

- **HA**: Traffic accidents in 2025 is less than 2024


You collect traffic data and run your statistical test. It returns a p-value of 0.076. To you reject or fail to reject your hypothesis, assuming $\alpha=0.05$?

Fail to reject null hypothesis

In words, explain what this *p-value* represents

The p-value of 0.076 means that if there was actually no decrease in traffic accidents between 2024 and 2025, there's a 7.6% chance we'd observe a difference this large just due to random variation. Since 7.6% is greater than our significance level of 0.05, we can't confidently conclude that the decrease is real rather than just random chance.
