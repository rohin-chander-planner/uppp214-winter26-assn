```python
import scipy
import pandas as pd
import numpy as np

df = pd.read_parquet(
    "https://github.com/oturns/example_datasets/raw/refs/heads/main/acs/ca_tracts_2021.pq"
)
df["county"] = df["geoid"].str[:5]
coastal_socal = ["06037", "06059", "06073"]
df_coastal = df[df.county.isin(coastal_socal)]

df_la = df[df.county == "06037"]
df_oc = df[df.county == "06059"]
df_sd = df[df.county == "06073"]
```

## Review

I am a regional planner considering implementing a rent-control policy in southern california. I'm concerned that OC may have a higher rent burden than LA county. How can I test this hypothesis?

Census tracts are designed to have roughly 4000 people. Are census tracts bigger (in population size) in OC than LA?

## ANOVA

### scipy

```python
F, p = scipy.stats.f_oneway(
    df_la.median_home_value,
    df_sd.median_home_value,
    df_oc.median_home_value,
    nan_policy="omit",
)
```

```python
F
```

```python
p
```

### pingouin

```python
# !pip install pingouin
```

```python
import pingouin as pg
aov = pg.anova(dv="median_home_value", 
                between="county", 
                data=df_coastal.dropna(subset=['median_home_value']), 
                detailed=True)
aov
```

### statsmodels

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

lm = ols("median_home_value ~ C(county)", data=df_coastal).fit()
lm.summary()
```

```python
sm.stats.anova_lm(lm)
```
